```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('features.csv')
display(df)
```

```python
# Define good (non-default = 0) and bad (default = 1)
df["good"] = 1 - df["default"]
df["bad"] = df["default"]
```

```python
df["age-bucket"] = pd.cut(df['age'], bins=[20, 35, 50, 70], labels = ["20-35", "36-50", "51-70"])
display(df)
```

```python
age_woe = df.groupby(['age-bucket'], observed=False)[['good', 'bad']].sum().reset_index(drop=False)
display(age_woe)
```

```python
age_woe['good dist'] = age_woe['good'] / age_woe['good'].sum()
age_woe['bad dist'] = age_woe['bad'] / age_woe['bad'].sum()
```

```python
age_woe['woe'] = np.log(age_woe['good dist'] / age_woe['bad dist'])
```
